练习:  创建两个DataFrame， 一个是期中成绩， 一个是期末成绩

```python
import numpy as np
import pandas as pd
from pandas import Series, DataFrame, Index
```

```python
# 学科： Java、Python、H5、MySQL、Oracle
# 人数： 100人
# 成绩的范围:  0 ~ 150
columns = ('Java', 'Python', 'H5', 'MySQL', 'Oracle')
data = np.random.randint(150, size=(100, 5))
df = DataFrame(data, columns=columns)
df.head()  # 显示前5条记录
```

```python
# 设置索引
df.index = ["A"+str(n) for n in range(1, 101)]
df
```

```python
# 显示第10位 ~ 第15位同学的各科成绩
# .loc[], df[]  行索引操作
df.loc['A10':'A15']  # 行索引的标签名
```

```python
df[9:15] # 行切片， 行索引值0开始
```

```python
df['A10':'A15']  # 行索引的标签
```

## df 操作的总结
- df['列名']  查看某一列
- df[['列名1', '列名2']]  查看多列
- df[:] 切片操作，只针对行索引，可以使用<b><font color='red'>索引值</font></b>或<b><font color='red'>索引名</font></b>

## .loc[]操作的总结
- df.loc['索引名'] 查看某一行
- df.loc[['索引名1', '索引名2']] 查看多行
- df.loc['索引名', '列名']  查看某一行的某一列
- df.loc['索引名', ['列名1', '列名2']] 查看某一行的多列
- df.loc['索引名1':'索引名2', '列名1':'列名2']  行和列的切片操作

## .iloc[] 操作的总结
- df.iloc[索引值,列位置] 第一部分是行索引， 第二部是列位置
- df.iloc[索引值1:索引值2]
- df.iloc[:,列位置1:列位置2]

```python
df.iloc[9:15,0:3]
```

# Series和DataFrame运算
- 使用函数方式
    - axis 表示按行或列进行运算 {0, 1, 'index', 'columns'}
        - 如果按行(0或'index')，两者之间依据行索引一一对应进行运算
        - 如果按列(1或'columns'),两者之间依据列一一对应进行运行
        - 说明：如果没有对应的行或列，则填充为NaN

如果两个DataFrame进行运算时，会自动地合并行或列(相同的)， 不同的行或列，会自动增加。

```python
java_python = df.head()[['Java', 'Python']]
h5 = df.head()['H5']
```

```python
display(java_python, h5)
```

```python
java_python.add(h5, axis=0)
```

```python
python_h5 = df.loc['A3':'A6',['Python', 'H5']]
python_h5
```

```python
df2 = java_python.add(python_h5, fill_value=0, axis=0)
df2
```

```python
python_s = Series([100, 20, 30, 80],
                  index=python_h5.index,
                  name='Python')
python_h5.add(python_s, axis='index')
```

如果 Series 和 DataFrame进行列运算时，Series的索引和DataFrame列名保持一致，如果不一致，则会新增列。一般用于对DataFrame的不同列进行不同数值的运算。

```python
# Python 50   -> 给Python列加上50分
# H5 10       -> 给H5列加上 10分
s = Series([50, 10],
                  index=python_h5.columns)

python_h5.add(s, axis='columns')
```

# DataFrame中的NaN值处理
- dropna()  删除Nan的数据
- fillna()  填充Nan的数据

```python
ddd2.loc[['张三', '蒋六'], ['语文', '英语']] = None
ddd2
```

```python
# axis {0 or 'index', 1 or 'columns'} 指定按行或列删除
# how {'any', 'all'} 
#     any 某一行或列中存在一个nan，则删除
#     all 某一行或列中所有元素都为nan时，则删除
ddd2.dropna(axis=0, how='all')
```

```python
ddd2.loc['disen'] = [np.nan]*3
ddd2
```

```python
ddd2.dropna(how='all')
```

```python
ddd2.fillna(value=1)  # 1) 指定值填充
```

```python
# 前置填充， 前一行的值填充后一行的Nan
ddd2.fillna(method='ffill', axis=0)  
```

```python
# 前置填充， 前一列的值填充后一列的nan
ddd2.fillna(method='ffill', axis=1)
```

```python
# 后置填充, 将后面的值向前填充Nan值
ddd2.fillna(method='bfill', axis=0)
```
