练习： 查看每一位同学的每门课的成绩
- join on 连接学生表student, 课程表 course

```python
sql="""
select s.sn, s.name sname, c.name cname, sc.score
from student s
join sc on (sc.sn = s.sn)
join course c on (sc.cn = c.cn)
"""
pd.read_sql(sql, conn, index_col='sn')
```

练习： 1、查询"01"课程比"02"课程成绩高的学生的信息及课程分数
- if 函数 + group by分组语句
- if(条件， True的值, False的值)

```python
sql = """
select s.sn 学号, s.name 姓名, A.c1 课程1, A.c2 课程2
from (
    select
        sn, 
        max(if(cn='01', score, 0)) c1,
        max(if(cn='02', score, 0)) c2
    from sc
    group by sn ) A
join student s on (A.sn = s.sn)
where A.c1 > A.c2 
"""

pd.read_sql(sql, conn)
```

```python
create_sql = """
create table c1_c2(
    学号 varchar(20),
    姓名 varchar(50),
    课程1 float,
    课程2 float
)
"""
cursor = conn.cursor()
```
