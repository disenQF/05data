```python
import numpy as np
import pandas as pd
from pandas import Series, DataFrame
import matplotlib.pyplot as plt
%matplotlib inline
```

## 线型图
- plot(kind='line')
- Series, DataFrame都可以绘制线型图

```python
# 生成Series, ndarray.cumsum() 数值累加
s = Series(np.random.randint(1, 20, size=100).cumsum()) 
```

```python
s.plot()
```

```python
df = DataFrame({
    'id': np.linspace(1, 50, num=50, dtype=np.uint8),
    'age': np.random.randint(18, 25, size=50),
    'salary': np.random.randint(6000, 30000, size=50)
})
df.head()
```

```python
# 将id列转成索引 
df.set_index('id', inplace=True)
```

```python
df.head()
```

```python
df['salary'].plot()
```

```python
ddd = DataFrame(np.random.uniform(1, 3, size=(10, 3)),
                columns=('白菜', '萝卜', '辣椒'),
                index=[ f'{month}月' for month in range(1, 11)])
ddd
```

```python
pd.__version__
```

```python
plt.rcParams['font.family'] = 'sans-serif'
plt.rcParams['font.sans-serif'] = ['KaiTi'] # KaiTi
```

```python
ddd.index.values
```

```python
ddd.plot(x=ddd.index)
```

```python
# 删除索引
ddd.reset_index(drop=True, inplace=True)
ddd.plot()
```

## 柱状图 bar/barh

```python
ddd.plot(kind='bar', legend=True)  # legend 图例是否显示
plt.legend(loc=2, shadow=True,
           facecolor='g',
           edgecolor='r') # 修改图例，loc是显示位置
```

```python
ddd.plot(kind='barh', legend=True)  # legend 图例是否显示
# loc = (0, 0) 左下角
# loc = (1, 1) 右上角
plt.legend(loc=(1.05, 0), shadow=True,
           facecolor='g',
           edgecolor='r') # 修改图例，loc是显示位置
```

## 直方图 hist
## 密度图 kde

```python
ddd.plot(kind='kde')
```

```python
ddd.plot(kind='hist')
```

## 散步图 
比较两个特征的数据，之间没有关系

```python
ddd.plot(x='白菜', y='萝卜',kind='scatter', c=['r', 'b'])
```

```python
df = DataFrame(np.random.randint(0,100,size = 100).reshape(50,2),  
                           columns = ['A','B'])

df.plot('A', 'B',
        kind = 'scatter', 
        title = 'x Vs y',
        c = ['r', 'b']
       )
```

```python
df = DataFrame(np.random.randn(200).reshape(50,4),
              columns=('A', 'B', 'C', 'D'))
df
```

```python
# 解决坐标轴上的负号问题
plt.rcParams['axes.unicode_minus'] = False
```

```python
# 默认diagonal = hist 直方图
pd.plotting.scatter_matrix(df,
                           alpha=1,
                           grid=True,
                           diagonal='kde',c='rybg')

plt.savefig('a.png')  # 保存画布
```
