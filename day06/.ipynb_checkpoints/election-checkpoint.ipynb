查看日期格式，并将其转换为Pandas的日期格式,通过函数加map方式进行转换

```python
from datetime import datetime

def format_date(item):
    # item是字符串的时间 01-AUG-11， 格式： %d-%b-%y
    # return datetime.strptime(item, '%d-%b-%y').strftime('%Y-%m-%d')
    return pd.to_datetime(item)
```

```python
election['contb_receipt_dt'] = election['contb_receipt_dt'].map(format_date)
```

把索引变成列,Series.reset_index()

找出各个候选人的捐赠者中，捐赠金额最大的人的职业以及捐献额  

通过query("查询条件来查找捐献人职业")

```python
nm_amt = DataFrame(election.groupby('cand_nm')['contb_receipt_amt'].max())
nm_amt.reset_index(inplace=True)
nm_amt.head()
```

```python
nm_amt.shape
```

```python
# 创建一个查询结果的DataFrame对象
result_df = DataFrame(columns=['cand_nm','contbr_occupation','contb_receipt_amt'])
```

```python
#捐赠金额最大的人的职业以及捐献额 
# contbr_occupation, contb_receipt_amt
for i in range(len(nm_amt)):
    row = nm_amt.iloc[i]  # 读取某一行数据，变为Series
    where = 'cand_nm == "{}" & contb_receipt_amt == {}'.format(row.iloc[0], row.iloc[1])
    row_d = election.query(where)[['cand_nm','contbr_occupation','contb_receipt_amt']]
    result_df = result_df.append(row_d)

# 显示查询的结果
result_df
```

```python
# query('查询表达式')  
# 表达式格式： 列名 == “条件值”  & 列名2 == “条件值2”
result = election.query('cand_nm=="Roemer, Charles E. \'Buddy\' III" & contb_receipt_amt==200.0')
result[['cand_nm','contbr_occupation','contb_receipt_amt']]
```

```python
# 合并方式查询
result2 = pd.merge(election, nm_amt)
result2[['cand_nm','contbr_occupation','contb_receipt_amt']]
```
