```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
a = np.random.randint(0, 255, size=(4,3))
a
```

```python
a[::-1]  # 将行反转过来
```

```python
a[:,::-1] # 将列反转过来
```

```python
a[::-1,::-1]  # 将行和列全部反转过来
```
